# Julia 機器學習：GLM 線性迴歸

## 作業 027：波士頓房價預測資料集

請使用 GLM 中的模型，建立一個預測模型來預測波士頓的房價。

```julia
using GLM, RDatasets, MLDataUtils
```

## 讀取資料

```julia
boston = dataset("MASS", "Boston")
first(boston, 10)
```

```julia
using StatsPlots
```

```julia
describe(boston)
```

```julia
gr(size = (600, 500))
@df boston corrplot([:MedV :Rm :Rad :Crim :Age :LStat :Tax], grid = false)
```

## 線性迴歸模型

### Crim

```julia
model = GLM.lm(@formula(MedV ~ Crim), boston)
```

```julia
intercept, slope = coef(model)
```

```julia
new_X = DataFrame(Crim=[5., 6., 7., 8., 9.])
```

```julia
predict(model, new_X)
```

```julia
GLM.r2(model)
```

```julia
GLM.adjr2(model)
```

### Rm

```julia
model = GLM.lm(@formula(MedV ~ Rm), boston)
```

```julia
intercept, slope = coef(model)
```

```julia
new_X = DataFrame(Rm=[5., 6., 7., 8., 9.])
```

```julia
predict(model, new_X)
```

```julia
GLM.r2(model)
```

```julia
GLM.adjr2(model)
```

## 多元線性迴歸模型

```julia
boston = dataset("MASS", "Boston")
first(boston, 10)
```

```julia
indecies = MLDataUtils.shuffleobs(collect(1:nrow(boston)))
train_ind, test_ind = MLDataUtils.splitobs(indecies, at = 0.8);
```

```julia
train = boston[train_ind, :]
test = boston[test_ind, :]
```

```julia
ols = GLM.lm(@formula(MedV ~ Crim + Zn + Indus + Chas + NOx + Rm + Age + Dis + Rad + Tax + PTRatio + Black + LStat), train)
```

```julia
predict(ols, test)
```

```julia
GLM.r²(ols)
```

```julia
GLM.adjr²(ols)
```
